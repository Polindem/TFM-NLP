## Detecci√≥n de descripci√≥n en ingl√©s

```python
import pandas as pd
import numpy as np
```

```python
movie_df = pd.read_csv('movies__Anna_traducido.csv')
movie_df.drop('Unnamed: 0', axis=1, inplace=True)
#movie_df.drop('Unnamed: 0.1', axis=1, inplace=True)
movie_df
```

```python
#dejamos caer las filas que contienen valores nulos
movie_df.dropna(subset = ["traduccion"], inplace=True)
#dejamos caer las columnas que contienen los datos que no vamos a usar
movie_df.drop(["descripcion","idioma"], inplace=True, axis=1)
movie_df
```

```python
%%time

# detectamos la traducci√≥n que ha quedado en ingl√©s

from langdetect import detect

lang = []    
    
for index, row in movie_df['traduccion'].iteritems():
    movie_df['traduccion'] = movie_df['traduccion'].apply(str)
    try:
        lang = detect(row) #detecting each row
    except:
        lang = "error"
    movie_df.loc[index, 'traduccion_idioma'] = lang

movie_df
```

```python
#pel√≠culas con la descripci√≥n en ingles
movie_df.loc[movie_df['traduccion_idioma'] == 'en']
```

```python
#pel√≠culas donde no se ha podido detectar el idioma
movie_df.loc[movie_df['traduccion_idioma'] == 'error']
```

```python
#dejamos caer las columnas que contienen los datos que no vamos a usar
movie_df.drop('Unnamed: 0.1', axis=1, inplace=True)
#dejamos caer las filas que contienen ingl√©s y error
movie_df = movie_df[movie_df.traduccion_idioma != 'en']
movie_df = movie_df[movie_df.traduccion_idioma != 'error']
movie_df
```

```python
# guardamos el dataset limpio
movie_df.to_csv('movies_final_Anna.csv', encoding='utf-8-sig')
```

### Juntamos tres datasets y analizamos los datos

```python
import glob 
import os
```

```python
# establecemos la ruta para unir varios archivos
files = os.path.join("movies_final_*.csv")
# lista de archivos combinados
files = glob.glob(files)
# unimos archivos con concat y read_csv
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
# eliminamos columna Unnamed: 0 para que no aparezca la numeraci√≥n individual de cada archivo
df.drop('Unnamed: 0', axis=1, inplace=True)
# guardamos el archivo conjunto
#df.to_csv('movies_final_all.csv', encoding='utf-8-sig')
```

```python
# analizamos la columna descripcion_idioma
df_detect_lang = df.groupby(['traduccion_idioma']).size().sort_values(ascending=False)
df_detect_lang
```

```python
# como hemos detectado m√°s idiomas en el dataset, creamos uno nuevo solamente con descripciones en espa√±ol
movies_final = df.loc[df['traduccion_idioma'] == 'es']
# eliminamos pel√≠culas sin descripci√≥n
movies_final = movies_final[movies_final.traduccion != "'Sin descripci√≥n'"]
movies_final.shape
```

```python
# guardamos el archivo
movies_final.to_csv('movies_final_all.csv', encoding='utf-8-sig')
```
